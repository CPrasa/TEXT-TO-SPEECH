```python
import tkinter as tk
from tkinter import *
from tkinter.ttk import Combobox
import pyttsx3

tts = pyttsx3.init()

def speak_now():
    
    text = text_box.get(1.0, END)
    gender = gender_box.get()
    speed = speed_box.get()
    voices = tts.getProperty('voices')

    def set_voice():
        if gender == 'Male':
            tts.setProperty('voice', voices[0].id)
            tts.say(text)
            tts.runAndWait()
        else:
            tts.setProperty('voice', voices[1].id)
            tts.say(text)
            tts.runAndWait()

    if text:
        if speed == 'Fast':
            tts.setProperty('rate', 200)
            set_voice()
        elif speed == 'Medium':
            tts.setProperty('rate', 120)
            set_voice()
        else:
            tts.setProperty('rate', 60)
            set_voice()

def stop_speech():
    tts.stop()

root = Tk()
root.title("CP Text to Speech Converter")
root.geometry("800x580+200+80")
root.resizable(False, False)
root.configure(bg="#B15EFF")

# Load the logo image
#cp_logo = PhotoImage(file="C:\\Users\\Asus\\OneDrive\\Desktop\\voiceApp\\images.png")

# Place the logo label
#Label(root, image=cp_logo, bg="#B15EFF").place(x=750, y=450)

# Set the application icon
logo_image = PhotoImage(file="C:\\Users\\Asus\\OneDrive\\Desktop\\voiceApp\\images.png")
root.iconphoto(False, logo_image)

# Create upper_frame
upper_frame = Frame(root, bg="#3D30A2", width=1000, height=100)
upper_frame.place(x=0, y=0)

# Add label with text
Label(upper_frame, text="Text to Speech Converter", font="TimesNewRoman 24 bold", bg="#3D30A2", fg='white').place(x=200, y=35)

text_box = Text(root, font="calibri 18", bg="#C2D9FF", relief=GROOVE, wrap=WORD, bd=0)
text_box.place(x=50, y=150, width=700, height=180)

# Creating a Combobox for gender selection
gender_box = Combobox(root, values=['Male', 'Female'], font="Roboto 12", state='readonly', width=12)
gender_box.place(x=200, y=400)
gender_box.set('Male')  # Set the default value

# Creating a Combobox for speed selection
speed_box = Combobox(root, values=['Fast', 'Medium', 'Slow'], font="Roboto 12", state='readonly', width=12)
speed_box.place(x=200, y=500)
speed_box.set('Medium')  # Set the default value

Label(root, text="Select Voice", font='TimesNewRoman 15 bold', bg="#B15EFF", fg="white").place(x=200, y=370)
Label(root, text="Select Speed", font='TimesNewRoman 15 bold', bg="#B15EFF", fg="white").place(x=200, y=470)

play_button = PhotoImage(file='play.png')

# Create the Play button
play_btn = Button(root, compound='left', image=play_button, bg='#B15EFF', width=130, height=100, font="arial 14 bold", borderwidth=0.1, command=speak_now)
#play_btn = Button(root, compound='left', image=play_button, bg='#B15EFF', width=130, height=100, font="arial 14 bold", borderwidth=0.1, command=lambda: play_btn.config(bg='#B15EFF') ,command=speak_now )
#play_btn = Button(root, compound='left', image=play_button, bg='#B15EFF', width=130, height=100, font="arial 14 bold", borderwidth=0.1, command=lambda: [play_btn.config(bg='#B15EFF'), speak_now])


play_btn.place(x=400, y=400)

# Create the Stop button
#stop_button = Button(root, text="Stop", font="arial 14 bold", bg='#FF5C5C', width=10, height=2, borderwidth=0.1, command=stop_speech)
#stop_button.place(x=300, y=475)

root.mainloop()
```
